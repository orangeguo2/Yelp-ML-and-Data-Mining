```python
df_review = pd.read_sql('SELECT * from review', connection)
df_user = pd.read_sql('SELECT * from user', connection)
df_business = pd.read_sql('SELECT * from business', connection)
```

```python
setStopWord = set(stopwords.words('english'))
def reviewPreProcessing(reviewSequ):
    Keepletters = re.sub('[^a-zA-Z]', ' ', reviewSequ)
    Keepletters = Keepletters.lower().split()
    removeWord = [x for x in Keepletters if not x in  setStopWord]
    return(' '.join(removeWord))
```

```python
def Text_pre(good_reviewCollection):
    result=[]
    for good_review in good_reviewCollection[0::200]:
        reviewsTokenized=TreebankWordTokenizer().tokenize(good_review)
        positiveTexts = []
        reviewsStemmed=""
        for text in good_review:
            positiveTexts += text      
        for word in reviewsTokenized:
            word=word.lower()
            reviewsStemmed+=WordNetLemmatizer().lemmatize(WordNetLemmatizer().lemmatize(word,'v'),'n') +" "
        ReviewsTemmedToken=TreebankWordTokenizer().tokenize(reviewsStemmed)
        ReviewsTemmedTokenStop=[txt for txt in ReviewsTemmedToken if not txt in setStopWord]
        for word in ReviewsTemmedTokenStop:
            if(word.isalpha()):
                result.append(word) 
    return result
```

```python
def Word_deal(good_reviewCollection):
    goodStr=""
    result = Text_pre(good_reviewCollection)
    print("result first step")
    print(result)
    PartTag=nltk.pos_tag(result)
    Taglen = len(PartTag)
    ReviewsSentimental = []
    for x in range(0,Taglen):
        if PartTag[x][0] in ['open','last','bad','horrible','new','awesome','different','first','terrible','good','great','excellent', 'many','sure','happy','nice','next','favorite','much','give','us', 'back', 'would','always', 'Las Vegas','get','like','little','small','would','really','also','got', 'go']:
            continue
        if(PartTag[x][1]=="JJ"):
            ReviewsSentimental.append(PartTag[x][0])
            goodStr+=str(PartTag[x][0])+" "
    return goodStr
```

```python
good_array = [5,4]
bad_array = [1,2]
good_reviewCollection = df_review.loc[df_review['stars'].isin(good_array)]['text']
bad_reviewCollection = df_review.loc[df_review['stars'].isin(bad_array)]['text']
```

```python
goodTextPrint= Word_deal(good_reviewCollection)
badTextPrint = Word_deal(bad_reviewCollection)
```

```python
stars = df_review['stars']
#reviewForSML=reviewPreProcessing(df_review['text'])
reviewForSML=df_review['text'].apply(reviewPreProcessing)
tfidfVector=TfidfVectorizer(ngram_range=(1,2)).fit_transform(reviewForSML)
```

```python
X_train, X_test, y_train, y_test = train_test_split(tfidfVector, stars, train_size=0.75)
```

```python
from sklearn.svm import LinearSVC
startTime = datetime.now()
clfSVC = LinearSVC()
#(kernel='linear', cache_size = 500)
clfSVC.fit(X_train, y_train)
preds_SVC = clfSVC.predict(X_test)

print('Accuracy Score: ')
print(accuracy_score(y_test, preds_SVC))

print('Classification Report: ')
print(classification_report(y_test, preds_SVC))

print('confusion_matrix: ')
print(confusion_matrix(y_test, preds_SVC))

print('The time used: ')
print(datetime.now()-startTime)
```
